```python
import sklearn
sklearn.__version__
```

```python
from sklearn.datasets import load_iris   # importando dataset iris do scikit learn
data = load_iris()
```

```python
import pandas as pd               # carregando dataaset iris como df do pandas
df = pd.DataFrame(data['data'], columns = data['feature_names'])
df['species'] = data['target']
df.head()
```

```python
df.plot()            # para criar um grafico de linhas com todas as variaveis df 
```

```python
df.plot.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)')  # scatter plot com 2 variaveis
```

```python
columns = ['sepal length (cm)', 'petal length (cm)', 'petal width (cm)', 'sepal width (cm)']  #graficos + complexos
df[columns].plot.area()
```

```python
df.groupby('species').mean().plot.bar()  # calcula a m das colunas agrup pela coluna spec e cria um graf c resultadof
```

```python
df.groupby('species').count().plot.pie(y = 'sepal length (cm)') # contagem de classes c grafuco de pizza
```

```python
df.plot.kde(subplots = True, figsize = (8,8)) #KDE para cada variavel do df 
```

```python
columns = ['sepal length (cm)', 'petal length (cm)', 'petal width (cm)', 'sepal width (cm)']   # boxplot de variavel num
df[columns].plot.box(figsize = (8,8))
```
