### Grafico de linha

```python

x = linspace(0, 5, 10)  #dados
y = x ** 2

fig = plt.figure()    #cria a figura

axes = fig.add_axes([0, 0, 0.8, 0.8])  # define a escala dos eixos

axes.plot(x, y, 'r')  # cria o plot

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de linha');
```

```python
x = linspace(0, 5, 10)  #dados
y = x ** 2

fig = plt.figure()    #cria a figura

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # eixos da figura principal 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  #eixos da figura secundária 

axes1.plot(x, y, 'r')  #figura principal
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

axes2.plot(y, x, 'g')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Secundária');

```

### Grafico de linhas em paralelo

```python
x = linspace(0, 5, 10)  #dados
y = x ** 2

fig, axes = plt.subplots(nrows = 1, ncols = 2)  #divide a area de plotagem pelos dois subplots

for ax in axes:           # loop pelos eixos para criar cada plot
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
    
fig.tight_layout()     # ajusta o layout
```

## Histogramas

```python
import numpy as np
```

```python
n = np.random.randn(100000)  # dados randomicos

fig, axes = plt.subplots(1, 2, figsize = (12, 4))  # cria subplots

axes[0].hist(n)            #plot1
axes[0].set_title("Histograma Padrão")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title("Histograma Cumulativo")   #plot2
axes[1].set_xlim((min(n), max(n)));
```

## Gráficos 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
alpha = 0.7   #dados
phi_ext = 2 * np.pi * 0.5 

def ColorMap(phi_m, phi_p):   # func para um mapa de cores
    return ( + alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X, Y = np.meshgrid(phi_p, phi_m)  # agrupa dados
Z = ColorMap(X, Y).T
```

```python
fig = plt.figure(figsize = (14,6))         

ax = fig.add_subplot(1, 2, 1,  projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0) #ajusta vaiáveis

ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False )

cb = fig.colorbar(p, shrink=0.5)                            
```

```python
#wire frame

fig = plt.figure(figsize=(8,6))    #cria a figura
ax = fig.add_subplot(1, 1, 1, projection ='3d')     #subplot
p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)   #wireframe
```
