## Criando gr√°fico com seaborn

```python
 dados = sea.load_dataset("tips")    # carregando um dos datasets que vem com o Seaborn
```

```python
dados.head()
```

```python
sea.jointplot(data = dados, x = "total_bill", y = "tip", kind = 'reg') # jointplot cria plot de 2 var com graficos bivariados e univariados
```

```python
sea.lmplot(data = dados, x = "total_bill", y = "tip", col = "smoker")
```

```python
df = pd.DataFrame()
```

```python
df['idade'] = random.sample(range(20, 100), 30)
df['peso'] = random.sample(range(55, 150), 30)
```

```python
df.shape
```

```python
df.head()
```

```python
sea.lmplot(data = df, x = "idade", y = "peso", fit_reg = True) 
```

```python
sea.kdeplot(df.idade)
```

```python
sea.kdeplot(df.peso)
```

```python
sea.distplot(df.idade)
```

```python
plt.hist(df.idade, alpha = .3)
sea.rugplot(df.idade)
```

```python
sea.boxplot(df.idade, color = 'm')
```

```python
sea.boxplot(df.peso, color = 'y')
```

```python
sea.violinplot(df.idade, color = 'm')
```

```python
sea.violinplot(df.peso, color = 'y')
```

```python
sea.clustermap(df)
```
